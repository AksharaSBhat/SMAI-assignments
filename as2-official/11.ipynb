```python
import numpy as np
def sigmoid_activation(vector):
    vector = 1 / (1 + np.exp(-vector))
    # vector = vector.T
    return vector

def sigmoid_activation_derivative(vector):
    vector = sigmoid_activation(vector)
    vector = vector * (1 - vector)
    # vector = vector.T
    return vector

def relu_activation(vector):
    vector = np.maximum(0,vector)
    # vector = vector.T
    return vector

def relu_activation_derivative(vector):
    vector = np.where(vector > 0, 1, 0)
    # vector = vector.T
    return vector

def tanh_activation(vector):
    vector = np.tanh(vector)
    # vector = vector.T
    return vector

def tanh_activation_derivative(vector):
    vector = 1 - np.tanh(vector) ** 2
    # vector = vector.T
    return vector
```

```python
import cupy as cp
import cv2
import numpy as np
import pandas as pd
import tqdm
import matplotlib.pyplot as plt

# ----- New: Data Loading and Preprocessing -----
def load_images_and_labels(csv_path, image_size=(28,28), symbol_map=None):
    """
    Loads images and constructs one-hot encoded labels.
    If symbol_map is not provided, it is created from the CSV.
    Returns:
        images: A CuPy array of shape (n_samples, input_dim)
        labels: A CuPy array of shape (n_samples, n_classes)
        symbol_map: The mapping from original symbol IDs to continuous labels.
    """
    df = pd.read_csv(csv_path)
    n_samples = len(df)
    input_dim = image_size[0] * image_size[1]

    # Create symbol map if not provided
    if symbol_map is None:
        unique_symbols = df['symbol_id'].unique()
        symbol_map = {s: i for i, s in enumerate(unique_symbols)}
    n_classes = len(symbol_map)

    images = np.empty((n_samples, input_dim), dtype=np.float32)
    labels = np.zeros((n_samples, n_classes), dtype=np.float32)

    for idx, row in df.iterrows():
        path = row['path']
        # Adjust path: remove first 6 characters and add "Symbols/" prefix
        path = path[6:]
        img = cv2.imread(path, 0)
        if img is None:
            # In case image reading fails, fill with zeros
            img = np.zeros(image_size, dtype=np.uint8)
        img = cv2.resize(img, image_size, interpolation=cv2.INTER_LANCZOS4)
        img = img.flatten().astype(np.float32) / 255.0
        images[idx] = img
        label_idx = symbol_map[row['symbol_id']]
        labels[idx, label_idx] = 1  # one-hot encoding

    # Convert to CuPy arrays
    images = cp.array(images).T  # shape becomes (input_dim, n_samples)
    labels = cp.array(labels).T  # shape becomes (n_classes, n_samples)
    return images, labels, symbol_map

# ----- New: Cross-Entropy Loss (with Softmax) -----
def softmax(z):
    # Subtract max for numerical stability
    z = z - cp.max(z, axis=0, keepdims=True)
    exp_z = cp.exp(z)
    return exp_z / cp.sum(exp_z, axis=0, keepdims=True)

def cross_entropy_loss(output, ground_truth):
    # Avoid log(0)
    eps = 1e-8
    return -cp.mean(cp.sum(ground_truth * cp.log(output + eps), axis=0))

def cross_entropy_loss_derivative(output, ground_truth):
    # When using softmax with cross entropy, the derivative simplifies to:
    return output - ground_truth

# ----- Revised Layer and Network Classes -----
class layer:
    def __init__(self, input_size, output_size,relu_activation=False):
        self.no_neurons_prev_layer = input_size
        self.no_neurons = output_size

        # self.weights = cp.random.randn(output_size, input_size) * cp.sqrt(2 / (input_size + output_size))
        self.weights = cp.random.randn(output_size, input_size).astype(cp.float16) * cp.sqrt(2 / (input_size + output_size))
        if relu_activation:
            self.weights = cp.random.randn(output_size, input_size) * cp.sqrt(2 / input_size)
        self.biases = cp.zeros((output_size, 1))
        # Placeholders for cache (for backprop)
        self.inputs = None
        self.weighted_input = None
        self.outputs = None

        # Gradients (accumulated over a batch)
        self.weights_gradient = cp.zeros((output_size, input_size))
        self.biases_gradient = cp.zeros((output_size, 1))
        self.deltas = cp.zeros((output_size, 1))

    def compute_output(self, activation_function, inputs):
        self.inputs = inputs  # shape: (input_size, batch_size)
        self.weighted_input = self.weights @ inputs + self.biases  # shape: (output_size, batch_size)
        self.outputs = activation_function(self.weighted_input)
        return self.outputs

    def compute_gradients(self, activation_derivative, deltas_nxt=None, weights_nxt=None, ground_truth=None):
        if ground_truth is not None:
            # Output layer: derivative of loss w.r.t weighted input
            # For softmax with cross entropy, gradient = output - ground_truth
            self.deltas = cross_entropy_loss_derivative(self.outputs, ground_truth)
        else:
            # Hidden layer
            self.deltas = (weights_nxt.T @ deltas_nxt) * activation_derivative(self.weighted_input)
        # Gradient for weights is outer product: delta @ inputs.T
        self.weights_gradient = self.deltas @ self.inputs.T
        self.biases_gradient = cp.sum(self.deltas, axis=1, keepdims=True)
        return self.deltas

    def update(self, learning_rate, batch_size):
        self.weights -= learning_rate * self.weights_gradient / batch_size
        self.biases -= learning_rate * self.biases_gradient / batch_size

class network:
    def __init__(self, input_size, output_size, hidden_layers, hidden_layer_size):
        self.hidden_layers = hidden_layers
        self.layers = []
        if hidden_layers == 0:
            self.layers.append(layer(input_size, output_size))
        else:
            self.layers.append(layer(input_size, hidden_layer_size))
            for _ in range(hidden_layers - 1):
                self.layers.append(layer(hidden_layer_size, hidden_layer_size))
            self.layers.append(layer(hidden_layer_size, output_size))

    def forward_pass(self, inputs, activation_function):
    # Hidden layers use the specified activation function.
      for lyr in self.layers[:-1]:
          inputs = lyr.compute_output(activation_function, inputs)
      # For the output layer, use an identity activation so that inputs get cached.
      # Then, apply softmax externally.
      linear_output = self.layers[-1].compute_output(lambda x: x, inputs)
      output = softmax(linear_output)
      # Optionally, update the output layer's cached outputs with the softmax output.
      self.layers[-1].outputs = output
      return output



    def train_batch(self, X, Y, activation_function, activation_derivative, learning_rate):
        # Forward pass
        output = self.forward_pass(X, activation_function)
        loss = cross_entropy_loss(output, Y)
        # Backpropagation: compute gradients layer by layer (backwards)
        # Output layer gradients
        delta = self.layers[-1].compute_gradients(activation_derivative, ground_truth=Y)
        # Hidden layers gradients
        for lyr in reversed(self.layers[:-1]):
            delta = lyr.compute_gradients(activation_derivative, deltas_nxt=delta, weights_nxt=self.layers[self.layers.index(lyr)+1].weights)
        # Update weights for all layers
        for lyr in self.layers:
            lyr.update(learning_rate, X.shape[1])
        return loss

    def predict(self, X, activation_function):
        output = self.forward_pass(X, activation_function)
        return cp.argmax(output, axis=0)

    def evaluate(self, X, Y, activation_function):
        preds = self.predict(X, activation_function)
        labels = cp.argmax(Y, axis=0)
        return cp.mean(preds == labels)

# ----- Revised Training Loop -----
def train_network(net, X_train, Y_train, epochs, learning_rate, activation_function, activation_derivative, batch_size):
    n_samples = X_train.shape[1]
    loss_history = []
    for epoch in tqdm.tqdm(range(epochs)):
        # Shuffle indices
        indices = cp.random.permutation(n_samples)
        X_train = cp.take(X_train, indices, axis=1)
        Y_train = cp.take(Y_train, indices, axis=1)

        epoch_loss = 0
        # Process mini-batches
        for i in range(0, n_samples, batch_size):
            X_batch = X_train[:, i:i+batch_size]
            Y_batch = Y_train[:, i:i+batch_size]
            loss = net.train_batch(X_batch, Y_batch, activation_function, activation_derivative, learning_rate)
            epoch_loss += loss * X_batch.shape[1]
        epoch_loss /= n_samples
        loss_history.append(epoch_loss.get())
        print(f"Epoch {epoch+1}/{epochs}, Loss: {epoch_loss.get():.4f}, activation: {activation_function.__name__}, batch size: {batch_size}")
        cp.get_default_memory_pool().free_all_blocks()
    return loss_history


activations_funcs = [relu_activation, sigmoid_activation, tanh_activation]
activations_derivatives = [relu_activation_derivative, sigmoid_activation_derivative, tanh_activation_derivative]


# Define list to store results from all activation and batch-size combinations
combinations_results = []

# Loop over activation functions and corresponding derivatives
for activation, activation_der in zip(activations_funcs, activations_derivatives):
    # For each activation, run for three batch configurations:
    for batch_label, batch_size_val in [('BGD', None),('MBGD', 1000), ('SGD', 32)]:
        fold_accs = []         # To store accuracy for each fold
        fold_loss_histories = []  # To store loss history (list of epoch losses) for each fold

        # Loop over folds
        for fold_no in range(1, 11):
            train_csv = f"classification-task/fold-{fold_no}/train.csv"
            test_csv  = f"classification-task/fold-{fold_no}/test.csv"

            # Load data
            X_train, Y_train, symbol_map = load_images_and_labels(train_csv)
            X_test, Y_test, _ = load_images_and_labels(test_csv, symbol_map=symbol_map)

            # Network parameters
            input_size = 28 * 28
            output_size = Y_train.shape[0]
            hidden_layers = 1
            hidden_layer_size = 500

            # Create network
            net = network(input_size, output_size, hidden_layers, hidden_layer_size)

            # Hyperparameters
            epochs = 10
            learning_rate = 0.1

            # Determine batch size:
            # If batch_size_val is None, we use the full training set (BGD)
            if batch_size_val is None:
                batch_size = X_train.shape[1]
            else:
                batch_size = batch_size_val

            # Train network and get the loss history over epochs
            loss_history = train_network(net, X_train, Y_train, epochs, learning_rate, activation, activation_der, batch_size)
            # Evaluate on test set
            accuracy = net.evaluate(X_test, Y_test, activation)

            fold_accs.append(accuracy.get())
            fold_loss_histories.append(loss_history)

        # Compute the average loss curve across folds (epoch-wise average)
        avg_loss_curve = np.mean(np.array(fold_loss_histories), axis=0)
        mean_acc = np.mean(fold_accs)
        std_acc = np.std(fold_accs)

        combinations_results.append({
            'activation': activation.__name__,
            'batch_type': batch_label,
            'mean_accuracy': mean_acc,
            'std_accuracy': std_acc,
            'avg_loss_curve': avg_loss_curve,
            'fold_accuracies': fold_accs,
        })

# Sort combinations in descending order of mean accuracy
sorted_results = sorted(combinations_results, key=lambda x: x['mean_accuracy'], reverse=True)

# Print the summary results and plot curves for each combination
for res in sorted_results:
    print(f"Activation: {res['activation']} | Batch Type: {res['batch_type']} | Mean Accuracy: {res['mean_accuracy']:.2f} | Std: {res['std_accuracy']:.2f}")

    # Plot the average loss curve
    plt.figure()
    plt.plot(res['avg_loss_curve'])
    plt.xlabel("Epoch")
    plt.ylabel("Average Loss")
    plt.title(f"Avg Loss Curve - {res['activation']} | {res['batch_type']}")
    plt.show()

    # Plot the accuracies for each fold
    plt.figure()
    plt.plot(res['fold_accuracies'], marker='o', linestyle='--')
    plt.xlabel("Fold")
    plt.ylabel("Accuracy")
    plt.title(f"Fold Accuracies - {res['activation']} | {res['batch_type']}")
    plt.show()


```

Mean Accuracy indicates the expected performance of the model on dataset, Higher the men acc, better the model
Standard deviation tells about the expected variablity in the performance of the model. Lower the std dev, more consistency ( and less overfitting)

Impact on Generalization Confidence:

Low Standard Deviation:

Increases confidence that the model will generalize well to unseen data.
Suggests that the models performance is stable across different subsets of data.
High Standard Deviation:

Reduces confidence in the models generalization since its performance is more unpredictable.
May indicate that the models performance could degrade significantly on new or slightly different data.

I would chosse the latter as Consistency (low standard deviation) implies that the model's performance is more reliable and robust across different samples. A higher variance means that the model might perform well on average but could fail in some cases, making it less trustworthy in real-world scenarios.
